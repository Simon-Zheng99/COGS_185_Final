```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
df = pd.read_csv("covid19_tweets.csv")
```

```python
df.head()
```

```python
text_data = df['text'].tolist()
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
sentiments = []
for text in text_data:
    sentiment_score = sia.polarity_scores(text)
    sentiment = max(sentiment_score, key=sentiment_score.get)
    sentiments.append(sentiment)
```

```python
df['sentiment'] = sentiments
```

```python
df.head()
```

```python
df["sentiment"].value_counts()
```

```python
sampled_df = df.sample(frac=0.2, random_state=42)
sampled_df
```

```python
#!pip install Xformers
from tqdm import tqdm
from transformers import pipeline

text_dataX = sampled_df['text'].tolist()

model_name = "cardiffnlp/twitter-roberta-base-sentiment"

sentiment_classifier = pipeline("sentiment-analysis", model=model_name)
```

```python
sentiment_labels = []
sentiment_scores = []
with tqdm(total=len(text_dataX), desc="Sentiment Analysis") as pbar:
    for text in text_dataX:
        sentiment = sentiment_classifier(text)
        sentiment_labels.append(sentiment[0]['label'])
        sentiment_scores.append(sentiment[0]['score'])
        pbar.update(1)
```

```python
sampled_df['sentiment_X'] = sentiment_labels
sampled_df['scores_X'] = sentiment_scores
```
