```python
import pandas as pd
```

```python
df = pd.read_csv('sampled_df.csv')
```

```python
df
```

```python
emotional_tweets = df[df['scores_X'] > 0.6]
emotional_tweets
```

```python
negative_tweets = emotional_tweets[emotional_tweets['sentiment_X'] == 'LABEL_0']
negative_tweets
```

```python
positive_tweets = emotional_tweets[emotional_tweets['sentiment_X'] == 'LABEL_2']
positive_tweets
```

```python
for text in negative_tweets['text']:
    text.split('\n')[0]
    
negative_tweets
```

```python
for index, row in negative_tweets.iterrows():
    text = row['text']
    modified_text = text.split('https')[0]
    negative_tweets.at[index, 'text'] = modified_text
negative_tweets
```

```python
negative_output = "negative_tweets.txt"

with open(negative_output, 'w') as file:
    for text in negative_tweets['text']:
        file.write(str(text) + '\n')
```

```python
for text in positive_tweets['text']:
    text.split('\n')[0]
    
positive_tweets
```

```python
for index, row in positive_tweets.iterrows():
    text = row['text']
    modified_text = text.split('https')[0]
    positive_tweets.at[index, 'text'] = modified_text
positive_tweets
```

```python
positive_output = "positive_tweets.txt"

with open(positive_output, 'w') as file:
    for text in positive_tweets['text']:
        file.write(str(text) + '\n')
```
